```python
# Get all users from Firestore
users = users_ref.get()

# Convert to list of dictionaries with IDs for easier reference
users_with_ids = [{"id": user.id, **user.to_dict()} for user in users]

# Randomly select 100 users to add collections to
selected_users = random.sample(users_with_ids, min(100, len(users_with_ids)))

for user in selected_users:
    user_doc_ref = users_ref.document(user["id"])

    # Create friends collection (2-8 random friends)
    num_friends = random.randint(2, 8)
    friends_ref = user_doc_ref.collection("friends")
    potential_friends = [u for u in users_with_ids if u["id"] != user["id"]]
    selected_friends = random.sample(
        potential_friends, min(num_friends, len(potential_friends))
    )

    for friend in selected_friends:
        friends_ref.add(
            {
                "user_id": friend["id"],  # Reference to the friend's user document
                "name": friend["name"],
                "email": friend.get("email", None),
                "friendship_date": datetime.datetime.now()
                - timedelta(days=random.randint(0, 365)),
            }
        )

    # Create children collection if age condition met
    user_age = user.get("age")
    if not user_age or user_age >= 20:
        children_ref = user_doc_ref.collection("children")
        num_children = random.randint(0, 3)

        for _ in range(num_children):
            children_ref.add(
                {
                    "name": f"{user['name'].split()[0]} Jr.",
                    "age": random.randint(1, max(1, user_age - 20) if user_age else 18),
                    "school": random.choice(
                        ["Elementary", "Middle", "High School", "College"]
                    ),
                }
            )

    # Create colleagues collection
    colleagues_ref = user_doc_ref.collection("colleagues")
    num_colleagues = random.randint(3, 6)
    potential_colleagues = [u for u in users_with_ids if u["id"] != user["id"]]
    selected_colleagues = random.sample(
        potential_colleagues, min(num_colleagues, len(potential_colleagues))
    )

    for colleague in selected_colleagues:
        colleagues_ref.add(
            {
                "user_id": colleague[
                    "id"
                ],  # Reference to the colleague's user document
                "name": colleague["name"],
                "email": colleague.get("email", None),
                "department": random.choice(
                    ["Engineering", "Sales", "Marketing", "HR", "Finance"]
                ),
            }
        )

    # Create neighbors collection
    neighbors_ref = user_doc_ref.collection("neighbors")
    num_neighbors = random.randint(2, 4)
    potential_neighbors = [u for u in users_with_ids if u["id"] != user["id"]]
    selected_neighbors = random.sample(
        potential_neighbors, min(num_neighbors, len(potential_neighbors))
    )

    for neighbor in selected_neighbors:
        neighbors_ref.add(
            {
                "user_id": neighbor["id"],  # Reference to the neighbor's user document
                "name": neighbor["name"],
                "email": neighbor.get("email", None),
                "house_number": f"{random.randint(1, 999)} {random.choice(['A', 'B', 'C', ''])}",
            }
        )

print(f"Added collections to {len(selected_users)} users")
```
